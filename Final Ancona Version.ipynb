```python
# Wind speed (resulting speed from U and V components)
ancona['wind_speed'] = np.sqrt(ancona['Wind-Speed (U)']**2 + ancona['Wind-Speed (V)']**2)

# Wind direction in radians
ancona['wind_dir_rad'] = np.arctan2(ancona['Wind-Speed (V)'], ancona['Wind-Speed (U)'])

# Convert to degrees from 0 to 360 (optional but nicer)
ancona['wind_dir_deg'] = (np.degrees(ancona['wind_dir_rad']) + 360) % 360
```

```python
ancona
```

```python
# Basic calendar features
ancona['hour'] = ancona['Date'].dt.hour
ancona['dayofweek'] = ancona['Date'].dt.dayofweek  # Monday=0, Sunday=6
ancona['month'] = ancona['Date'].dt.month
ancona['is_weekend'] = ancona['dayofweek'] >= 5
```

```python
def month_to_season(m):
    if m in [12, 1, 2]:
        return 0  # winter
    elif m in [3, 4, 5]:
        return 1  # spring
    elif m in [6, 7, 8]:
        return 2  # summer
    else:
        return 3  # autumn

ancona['season'] = ancona['month'].apply(month_to_season)
```

```python
ancona
```

```python
# Cyclic hour encoding:
# hour is periodic (after 23 comes 0), so we map it onto a circle using:
# sin(2π * hour/24) and cos(2π * hour/24).
# This helps models learn daily patterns without a false jump between 23 and 0.
```

```python
ancona['hour_sin'] = np.sin(2 * np.pi * ancona['hour'] / 24)
ancona['hour_cos'] = np.cos(2 * np.pi * ancona['hour'] / 24)
```

```python
max_lag = 24  # use last 24 hours

for lag in range(1, max_lag + 1):
    ancona[f'PM25_lag_{lag}'] = ancona.groupby('station_name')['PM2.5'].shift(lag)
```

```python
for window in [3, 6, 12]:
    ancona[f'PM25_roll_mean_{window}h'] = (
        ancona.groupby('station_name')['PM2.5']
              .transform(lambda s: s.rolling(window).mean())
    )
    ancona[f'PM25_roll_max_{window}h'] = (
        ancona.groupby('station_name')['PM2.5']
              .transform(lambda s: s.rolling(window).max())
    )
    ancona[f'PM25_roll_min_{window}h'] = (
        ancona.groupby('station_name')['PM2.5']
              .transform(lambda s: s.rolling(window).min())
    )
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

best_params = {
    'n_estimators': 200,
    'min_samples_split': 5,
    'min_samples_leaf': 1,
    'max_features': 0.7,
    'max_depth': 25
}

def print_metrics(y_true, y_pred, name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{name}  MAE = {mae:.3f}, RMSE = {rmse:.3f}")
```

```python
results = []

for st in stations:
    # Filter data for this station
    train_st = train_data[train_data['station_name'] == st]
    val_st   = val_data[val_data['station_name'] == st]
    test_st  = test_data[test_data['station_name'] == st]
    
    # In case some station has no data in a split, skip it
    if len(train_st) == 0 or len(val_st) == 0 or len(test_st) == 0:
        continue
    
    X_train_st = train_st[feature_cols]
    y_train_st = train_st['PM2.5']
    
    X_val_st = val_st[feature_cols]
    y_val_st = val_st['PM2.5']
    
    X_test_st = test_st[feature_cols]
    y_test_st = test_st['PM2.5']
    
    # Create and fit station-specific RF
    rf_local = RandomForestRegressor(
        **best_params,
        random_state=42,
        n_jobs=-1
    )
    rf_local.fit(X_train_st, y_train_st)
    
    # Predictions
    y_val_pred_st  = rf_local.predict(X_val_st)
    y_test_pred_st = rf_local.predict(X_test_st)
    
    # Metrics
    mae_val  = mean_absolute_error(y_val_st, y_val_pred_st)
    rmse_val = np.sqrt(mean_squared_error(y_val_st, y_val_pred_st))
    
    mae_test  = mean_absolute_error(y_test_st, y_test_pred_st)
    rmse_test = np.sqrt(mean_squared_error(y_test_st, y_test_pred_st))
    
    results.append({
        'station': st,
        'val_mae': mae_val,
        'val_rmse': rmse_val,
        'test_mae': mae_test,
        'test_rmse': rmse_test
    })
    
    print(f"Station: {st}")
    print(f"  Val  MAE={mae_val:.3f}, RMSE={rmse_val:.3f}")
    print(f"  Test MAE={mae_test:.3f}, RMSE={rmse_test:.3f}")
    print("-" * 40)

results_df = pd.DataFrame(results)
results_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- Decision Tree pipeline ---
dt_model = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", DecisionTreeRegressor(
        random_state=42,
        max_depth=10
    ))
])

# Fit on training data
dt_model.fit(X_train, y_train)

# Predict on validation and test
y_val_pred  = dt_model.predict(X_val)
y_test_pred = dt_model.predict(X_test)

# Metrics
def regression_metrics(y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    return mae, rmse, r2

val_mae, val_rmse, val_r2 = regression_metrics(y_val, y_val_pred)
test_mae, test_rmse, test_r2 = regression_metrics(y_test, y_test_pred)

results_dt = pd.DataFrame({
    "Split": ["Validation", "Test"],
    "MAE": [val_mae, test_mae],
    "RMSE": [val_rmse, test_rmse],
    "R2": [val_r2, test_r2]
})

results_dt
```
